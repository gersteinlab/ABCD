# XGBOOST

```python
#xgboost
seed = [i for i in range(9300, 9310)]
out = {}
for s in seed:
    trainer = ABCDTrainer(config)
    trainer.fit_sklearn(X_out, Y, 1, ablation = True, ablation_target = ['C1', 'C2', 'cog', 'C4_1','C4_2','C4_3','C4_4','C4_5','C4_6','C4_7'], random_state = s)
    if s == 9300:
        [print(o) for o in trainer.feature_names]
        out['Feature_names'] = trainer.feature_names
    [print(o) for o in trainer._learn.feature_importances_]
    out[s] = trainer._learn.feature_importances_
```

```python
pd.DataFrame(out).to_csv('out.csv')
```
